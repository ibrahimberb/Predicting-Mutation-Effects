```python
def visualize_label_counts(data_param, label_name_param="Mutation_Effect_Label"):
    # noinspection SpellCheckingInspection
    sns.set(style="white", font_scale=1.15)  # white, dark, whitegrid, darkgrid, ticks
    val_counts = data_param[label_name_param].value_counts().sort_index()
    val_counts = val_counts.rename({0: 'Disrupting', 1: 'Increasing + No Effect'})
    print(f"Label counts:\n{val_counts}")
    ax = sns.barplot(x=val_counts.index,
                     y=val_counts,
                     palette="ch:s=-.2,r=.6")
    ax.set_title('Disrupting vs Increasing & No Effect')  # ch:s=-.2,r=.6, ocean
    ax.set_ylabel('Value counts')
```

```python
concated = pd.concat(
    [
        brca_data["Prediction"].value_counts(),
        esca_data["Prediction"].value_counts(),
        gbm_data["Prediction"].value_counts(),
        hnsc_data["Prediction"].value_counts(),
        ov_data["Prediction"].value_counts()
    ], axis=1
)
concated.columns = ["BRCA", "ESCA", "GBM", "HNSC", "OV"]
concated.index.name = "Mutation_Effect"
concated = concated.reset_index()
concated
```

```python
L = []
for i in range(5):
    L.append("Disrupting")
    L.append("Increasing+NoEffect")
L
```

```python
concated_final = concated[["BRCA", "ESCA", "GBM", "HNSC", "OV"]].melt(var_name="TCGA Cohorts",
                                                                      value_name="# of Interactions")
concated_final.insert(0, "Mutation_Effect", L)
concated_final
```

```python
plt.figure(figsize=(10, 8), dpi=600)
sns.set_theme(style="white", palette="Set1", font_scale=1.30)  # Set2, Accent
plt.rc('font', family='serif')
sns.barplot(x="TCGA Cohorts", y="# of Interactions", hue="Mutation_Effect", data=concated_final)
sns.barplot(x="TCGA Cohorts", y="# of Interactions", hue="Mutation_Effect", data=concated_final)
plt.show()
```

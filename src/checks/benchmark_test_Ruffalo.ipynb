```python
import pandas as pd
```

```python
BRCA_PATIENT_INTERACTION_DATA_PATH = r"C:\Users\ibrah\Documents\GitHub\Predicting-Mutation-Effects\data\patient_interaction_datasets\BRCA_patient_interactions_analysis_table_2021-11-17.xlsx"
```

```python
brca_patient_interaction_data = pd.read_excel(BRCA_PATIENT_INTERACTION_DATA_PATH)
brca_patient_interaction_data.head()
```

```python
# The changes will be on this data frame
brca_patient_interaction_data_prepared = brca_patient_interaction_data.copy()

# Drop entries if DISRUPTIVE_INTERACTORS is nan
brca_patient_interaction_data_prepared.dropna(subset=["DISRUPTIVE_INTERACTORS"], inplace=True)

# Extract the gene values on another column
brca_patient_interaction_data_prepared["GENE"] = brca_patient_interaction_data_prepared["PROTEIN_GENE"].apply(lambda x: x.split(":")[1])

# Get rid of unnecessary columns
brca_patient_interaction_data_prepared = brca_patient_interaction_data_prepared[["PATIENT", "GENE", "DISRUPTIVE_INTERACTORS"]].copy()

# split the protein_gene values so that each cell contains a list
brca_patient_interaction_data_prepared["DISRUPTIVE_INTERACTORS"] = brca_patient_interaction_data_prepared["DISRUPTIVE_INTERACTORS"].apply(lambda x: x.split(","))

# explode over the values in the list in the cells
brca_patient_interaction_data_prepared = brca_patient_interaction_data_prepared.explode("DISRUPTIVE_INTERACTORS")

# Extract the interactor gene values on DISRUPTIVE_INTERACTORS column.
brca_patient_interaction_data_prepared["DISRUPTIVE_INTERACTORS"] = brca_patient_interaction_data_prepared["DISRUPTIVE_INTERACTORS"].apply(lambda x: x.split(":")[1])

# final version
brca_patient_interaction_data_prepared
```

```python
# The changes will be on this data frame
brca_patient_interaction_data_prepared_all_interactors = brca_patient_interaction_data.copy()

# Extract the gene values on another column
brca_patient_interaction_data_prepared_all_interactors["GENE"] = brca_patient_interaction_data_prepared_all_interactors["PROTEIN_GENE"].apply(lambda x: x.split(":")[1])

# Get rid of unnecessary columns
brca_patient_interaction_data_prepared_all_interactors = brca_patient_interaction_data_prepared_all_interactors[["PATIENT", "GENE", "INTERACTORS"]].copy()

# split the protein_gene values so that each cell contains a list
brca_patient_interaction_data_prepared_all_interactors["INTERACTORS"] = brca_patient_interaction_data_prepared_all_interactors["INTERACTORS"].apply(lambda x: x.split(","))

# explode over the values in the list in the cells
brca_patient_interaction_data_prepared_all_interactors = brca_patient_interaction_data_prepared_all_interactors.explode("INTERACTORS")

# Extract the interactor gene values on INTERACTOR column.
brca_patient_interaction_data_prepared_all_interactors["INTERACTORS"] = brca_patient_interaction_data_prepared_all_interactors["INTERACTORS"].apply(lambda x: x.split(":")[1])

# final version
brca_patient_interaction_data_prepared_all_interactors
```
